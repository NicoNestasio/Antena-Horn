# 13GHz

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Ruta del archivo Excel
ruta_csv = '/content/13GHz - Pol. Principal - Diagrama de radiacin-raw.csv'

# Importar datos desde Excel a un DataFrame de pandas
datos = pd.read_csv(ruta_csv)
datos['Positions'] = datos['Positions'].str.replace('deg', '').astype(float)
datos['Amplitudes'] = datos['Amplitudes'].str.replace('dBm', '').astype(float)

# Ver los primeros registros para asegurarte de que los datos se importaron correctamente
print(datos.head())
print(datos.columns)

# Acceder a los datos de las columnas
angulos = np.deg2rad(datos['Positions'])
amplitudes = datos['Amplitudes']


# Encontrar el punto mximo
maximo_idx = np.argmax(amplitudes)
angulo_max = angulos[maximo_idx]
amplitud_max = amplitudes[maximo_idx]
diferencia = 13.07-amplitud_max

# Correccin de datos
amplitudescor = amplitudes+diferencia
amplitud_maxcor= 13.07
anguloscor = angulos - angulo_max
# Crear el grfico
plt.figure()
plt.polar(anguloscor, amplitudescor, marker='o', markersize=1, label='Radiacin')
plt.polar(0, amplitud_maxcor, 'ro', label='Mximo')
plt.title('Diagrama de Radiacin Medicin (dB)')
plt.legend()
plt.ylim(-15, 15)
plt.yticks(np.arange(-15, 16, 5))
plt.show()
print('Mximo:',amplitud_maxcor)
```

```python
# Grfico (Simulacin)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Ruta del archivo Excel
ruta_csv4 = '/content/Diagrama 13GHz.csv'

# Importar datos desde Excel a un DataFrame de pandas
datos4 = pd.read_csv(ruta_csv4)
datos4.rename(columns={datos4.columns[1]: 'dB(GainTotal)'}, inplace=True)
# Ver los primeros registros para asegurarte de que los datos se importaron correctamente
print(datos4.head())
print(datos4.columns)

# Acceder a los datos de las columnas
angulos4 = np.deg2rad(datos4['Phi [deg]'])
amplitudes4 = datos4['dB(GainTotal)']


# Encontrar el punto mximo
maximo_idx4 = np.argmax(amplitudes4)
angulo_max4 = angulos4[maximo_idx4]
amplitud_max4 = amplitudes4[maximo_idx4]
dif_max=amplitud_max4-amplitud_maxcor

# Crear el grfico
plt.figure(figsize=(10, 20))
plt.polar(angulos4, amplitudes4, marker='o', markersize=0, label='Simulacin')
plt.polar(angulo_max4, amplitud_max4, 'ro', label='Mximo')
plt.polar(anguloscor, amplitudescor, marker='o', markersize=1, label='Medicin')
plt.polar(0, amplitud_maxcor, 'go', label='Mximo')
plt.title('Diagrama de Radiacin Simulacin(dB)')
plt.legend()
plt.ylim(-15, 18)  # Por ejemplo, establece el mnimo en 0 y el mximo en 20 dB
plt.yticks(np.arange(-15, 16, 5))
plt.show()
plt.show()
print('Mximo Simulado:',amplitud_max4 )
print('Mximo Medido:',amplitud_maxcor )
print('Diferencia:', dif_max)
```
